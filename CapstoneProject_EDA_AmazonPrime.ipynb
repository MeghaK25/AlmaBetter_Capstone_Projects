## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# creating copies of original dataset
titles_clean = titles_df.copy()
credits_clean = credits_df.copy()

# Handling duplicate records
titles_clean = titles_clean.drop_duplicates()
credits_clean = credits_clean.drop_duplicates()
````

````python
# Handle Missing Values – Titles Dataset

# Seasons (Movies do not have seasons)
titles_clean['seasons'] = titles_clean['seasons'].fillna(0)

# Age Certification (Fill missing as 'Unknown')
titles_clean['age_certification'] = titles_clean['age_certification'].fillna('Unknown')

# Description
titles_clean['description'] = titles_clean['description'].fillna('No Description Available')
````

````python
# Handle Missing Values – Credits Dataset

# Character Column (Missing for Directors)
credits_clean['character'] = credits_clean['character'].fillna('Not Applicable')
````

````python
# Convert Data Types - Seasons as Integer
titles_clean['seasons'] = titles_clean['seasons'].astype(int)

# Standardize Text Columns
titles_clean['type'] = titles_clean['type'].str.upper()
credits_clean['role'] = credits_clean['role'].str.upper()

# Split Multi-Value Columns for Easier Analysis

#############
# Clean and standardize genres column
titles_clean['genres_clean'] = (
    titles_clean['genres']
    .str.replace('[', '', regex=False)
    .str.replace(']', '', regex=False)
    .str.replace("'", '', regex=False)
    .str.lower()
    .str.split(',')
)

# Final cleaning of genres after splitting
titles_clean['genres_clean'] = titles_clean['genres_clean'].apply(
    lambda x: [g.strip().lower() for g in x if g and g.strip() != '']
)

############
# Genres
titles_clean['genres_list'] = titles_clean['genres'].str.split(',')

# Production Countries
titles_clean['production_countries_list'] = titles_clean['production_countries'].str.split(',')
````

Scatter plot is appropriate to detect trends, clustering, and diminishing returns between two numerical variables. To analyze the relationship between director productivity (number of titles directed) and audience popularity (TMDb popularity).


##### 2. What is/are the insight(s) found from the chart?

* Directors with low frequency (2–5 titles) often achieve the highest average TMDb popularity.

* As the number of titles directed increases, average popularity drops sharply and stabilizes at low values.

* Highly prolific directors (15+ titles) rarely appear in the high-popularity region.

* Popularity seems driven by selective, high-impact projects, not volume.
